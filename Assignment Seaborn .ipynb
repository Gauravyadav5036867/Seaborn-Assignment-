```python
#Question 1

Seaborn is a powerful library for statistical data visualization in Python, built on top of Matplotlib. Here are five common plots you can create using Seaborn along with their typical uses:

Scatter Plot

Use: Visualize the relationship between two numerical variables. Useful for identifying correlations or clusters in data.
Histogram

Use: Display the distribution of a single numerical variable. Helps in understanding data spread, central tendency, and skewness.
Bar Plot

Use: Show comparisons among categorical variables or between categorical and numerical variables. Useful for displaying counts or aggregations.
Box Plot

Use: Summarize the distribution of numerical data through quartiles. It shows outliers and skewness in data effectively.
Heatmap

Use: Visualize data in a matrix format where values are represented by colors. Useful for showing correlations or distributions across two categorical variables.
```

```python
#Question 2

To plot a line plot using the "fmri" dataset from Seaborn, where we plot timepoint on the x-axis and signal on the y-axis for different events and regions, we can proceed with the following steps:

Load the "fmri" dataset using Seaborn's load_dataset function.
Use Seaborn's lineplot function to plot timepoint vs signal with event as the hue to differentiate lines by different events.
Optionally, use region as a separate plot for different regions.


import seaborn as sns
import matplotlib.pyplot as plt


fmri_data = sns.load_dataset("fmri")

plt.figure(figsize=(10, 6))
sns.lineplot(x="timepoint", y="signal", hue="event", data=fmri_data, ci=None)




plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("FMRI Signal Over Time by Event")
plt.legend(title="Event", loc="upper right")

plt.show()
```

```python
#Question 3

import seaborn as sns
import matplotlib.pyplot as plt


titanic_data = sns.load_dataset("titanic")


plt.figure(figsize=(12, 6))


plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic_data)
plt.title('Age distribution across passenger classes')
plt.xlabel('Pclass')
plt.ylabel('Age')


plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic_data)
plt.title('Fare distribution across passenger classes')
plt.xlabel('Pclass')
plt.ylabel('Fare')

plt.tight_layout()
plt.show()
```

```python
#Question 4

import seaborn as sns
import matplotlib.pyplot as plt


diamonds_data = sns.load_dataset("diamonds")

plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', bins=30, kde=True)
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Count')
plt.legend(title='Cut', loc='upper right')

plt.show()
```

```python
#Question 5

import seaborn as sns
import matplotlib.pyplot as plt


iris_data = sns.load_dataset("iris")


sns.pairplot(iris_data, hue='species', height=2.5)

plt.suptitle("Pair Plot of Iris Dataset with Species Differentiation", y=1.02)
plt.show()
```

```python
#Question 6

import seaborn as sns
import matplotlib.pyplot as plt


flights_data = sns.load_dataset("flights")


flights_pivot = flights_data.pivot(index='month', columns='year', values='passengers')

plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, annot=True, cmap='YlGnBu', fmt='d', linewidths=.5)
plt.title('Heatmap of Flight Passengers')
plt.xlabel('Year')
plt.ylabel('Month')

plt.show()
```
